```python
#Importing modules
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Defining axis range
x = np.array(range(100))
y = x
```

```python
# Create the plot

#오른 상단 버튼 클릭시 레이블 선택기능
plt.plot(x,y**2.2, label = "good economy")

plt.plot(x,y**2, label = "medium economy")
plt.plot(x,y**1.9, label = "fucked ecomony")

# 타이틀 버튼을 누르고 아래 코드 실행
plt.title("Simple Function Visualization Example")

# Add X and y Label
plt.xlabel('CEO')
plt.ylabel('Stock')

# Add a grid; aplpha defined as grid thickness
plt.grid(alpha=.6,linestyle='-')

#add legend
plt.legend()


####x1 0 -100;   x2 0 - 100    
```
